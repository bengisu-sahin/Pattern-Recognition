```python
# RGB2YIQ FUNCTION
def convertRGB2YIQ(task1_img:np.array,conversion_matrix:np.array):
    # Resmin boyutlarn al
    height, width, _ = task1_img.shape

    # YIQ'ye dnm iin bo bir dizi olutur
    img_yiq = np.zeros(task1_img.shape)

    #rgb ve conversation matrisini arparak yiq grntsn hesapla
    for i in range(0,height):
      for j in range(0,width):
        img_yiq[i,j, ]=np.dot(conversion_matrix,task1_img[i,j, ])
    return img_yiq
```

```python
# YIQ2RGB FUNCTION
def convertYIQ2RGB(task1_img:np.array,conversion_matrix:np.array):
    # Resmin boyutlarn al
    height, width, _ = task1_img.shape

    # YIQ'ye dnm iin bo bir dizi olutur
    img_yiq = np.zeros(task1_img.shape)

    #rgb ve conversation matrisini arparak yiq grntsn hesapla
    for i in range(0,height):
      for j in range(0,width):
        img_yiq[i,j, ]=np.dot(conversion_matrix,task1_img[i,j, ])
    return img_yiq
```

```python
# CONVERT RGB2YIQ COLOR TEMPLATE IMAGE
color_template=cv2.cvtColor(cv2.imread('/content/color_template.png'), cv2.COLOR_BGR2RGB)
# Dnm matrisi
conversion_matrix_to_yiq = np.array([[+0.2126, +0.7152, +0.0722],
                                    [-0.09991, -0.33609, 0.436],
                                    [+0.615, -0.55861, -0.05639]], dtype=float)

img_yiq=convertRGB2YIQ(color_template,conversion_matrix_to_yiq)
plt.title('RGB to YIQ Image')
plt.imshow(img_yiq.astype(np.uint8))
plt.show()


# CONVERT YIQ2RGB COLOR TEMPLATE IMAGE
conversion_matrix_to_rgb= np.array([[+1, 0, 1.28033],
                        [+1, -0.21482, -0.38059],
                        [+1, 2.12798, 0]], dtype=float)
img_rgb=convertYIQ2RGB(img_yiq,conversion_matrix_to_rgb)
plt.title('YIQ to RGB Image')
plt.imshow(img_rgb.astype(np.uint8))
plt.show()
```

### TASK 2

```python
import cv2
from matplotlib import pyplot as plt

# Kedi resimlerinin dosya adlarn bir listede topla
images = ['/content/cat1.jpg', '/content/cat2.jpg', '/content/cat3.jpg','/content/dog1.jpg']

# Her bir resim zerinde dng
for i, image_path in enumerate(images, start=1):
    # Resmi oku
    image = cv2.imread(image_path)

    # Yatay ve dikey evirme
    image_vflip = cv2.flip(image, 0)
    image_hflip = cv2.flip(image, 1)

    # Saa ve sola dnme
    image_rotate_left = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
    image_rotate_right = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

    # Boyutu yarya indirme
    image_resized = cv2.resize(image, (int(image.shape[1] / 2), int(image.shape[0] / 2)))

    # Grselletirme
    plt.figure(figsize=(10, 8))

    plt.subplot(2, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {i} Original {image.shape}')

    plt.subplot(2, 3, 2)
    plt.imshow(cv2.cvtColor(image_vflip, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {i} Vertical Flip')

    plt.subplot(2, 3, 3)
    plt.imshow(cv2.cvtColor(image_hflip, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {i} Horizontal Flip')

    plt.subplot(2, 3, 4)
    plt.imshow(cv2.cvtColor(image_rotate_left, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {i} Rotate Left')

    plt.subplot(2, 3, 5)
    plt.imshow(cv2.cvtColor(image_rotate_right, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {i} Rotate Right')

    plt.subplot(2, 3, 6)
    plt.imshow(cv2.cvtColor(image_resized, cv2.COLOR_BGR2RGB))
    plt.title(f'Image {i} Resized {image_resized.shape}')

    plt.tight_layout()
    plt.show()
```
